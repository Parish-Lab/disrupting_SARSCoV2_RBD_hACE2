These calculations are preliminary and should be verified by watching the trajectories and comparing the frames where the ligand is outside the bindng site (found by this calculation) with the frames viewed.

Using MM-GBSA binding free energy to indicate when the current ligand is not bound to anything and then using COM to see how often it leaves the binding site (10 angstroms in this instance)

```python
#Calculating the percent dissociation for each ligand and corresponding poses

#Using MM-GBSA binding free energy to indicate when the current ligand is not bound to anything
#and then using COM to see how often it leaves the binding site (10 angstroms)

###KEEPING TRACK OF THE OUT FRAMES
#Create a list that will hold the frames where the ligand is considered outside of the binding site for every ligand/pose.
#This will now be referred to as "out frames"
All_Out_Frames = []
#Out frames due to MMGBSA 
AOF_MMGBSA = []
#Out frames due to center of mass
AOF_COM = []

#Number of frames of each lifand-bound simulation
num_frames_MD_sims = 5000


#Create a list that will hold the percent dissociation for every ligand/pose
Final_PD = []

###INPUT FILES AND MMGBSA PLOT INPUT FILES
#Create a list that contains the name of all the MM-GBSA text files, MM_GBSA csv files and Center of Mass (COM) text files
#Will create MM_GBSA text files with only the average MM-GBSA binding estimations (in the future can create a plot)
MMGBSA_txt = ['MMGBSA_Diquafosol_P1.txt', 'MMGBSA_Diquafosol_P3.txt', 'MMGBSA_Emodin_P1.txt', 'MMGBSA_Emodin_P2.txt', 'MMGBSA_Fosinopril_P1.txt', 'MMGBSA_Fosinoprilat_P1.txt', 'MMGBSA_Fosinoprilat_P2.txt', 'MMGBSA_Fosinoprilat_P3.txt', 'MMGBSA_Lisinopril_P1.txt', 'MMGBSA_Lisinopril_NH_P1.txt', 'MMGBSA_Physcion_P1.txt', 'MMGBSA_Physcion_P2.txt', 'MMGBSA_Physcion_P3.txt']
MMGBSA_csv = ['Delta_MMGBSA_Lig_Specific_DBind_6lzg_COVID_hACE2_glycans_Zn_Diquafosol_P1_BSite_100ns_All_Seeds.csv', 'Delta_MMGBSA_Lig_Specific_DBind_6lzg_COVID_hACE2_glycans_Zn_Diquafosol_P3_BSite_100ns_All_Seeds.csv', 'Delta_MMGBSA_Lig_Specific_DBind_6lzg_COVID_hACE2_glycans_Zn_Emodin_P1_BSite_100ns_All_Seeds.csv', 'Delta_MMGBSA_Lig_Specific_DBind_6lzg_COVID_hACE2_glycans_Zn_Emodin_P2_BSite_100ns_All_Seeds.csv', 
              'Delta_MMGBSA_Lig_Specific_DBind_6lzg_COVID_hACE2_glycans_Zn_Fosinopril_P1_BSite_100ns_All_Seeds.csv', 'Delta_MMGBSA_Lig_Specific_DBind_6lzg_COVID_hACE2_glycans_Zn_Fosinoprilat_P1_BSite_100ns_All_Seeds.csv', 'Delta_MMGBSA_Lig_Specific_DBind_6lzg_COVID_hACE2_glycans_Zn_Fosinoprilat_P2_BSite_100ns_All_Seeds.csv', 'Delta_MMGBSA_Lig_Specific_DBind_6lzg_COVID_hACE2_glycans_Zn_Fosinoprilat_P3_BSite_100ns_All_Seeds.csv', 
              'Delta_MMGBSA_Lig_Specific_DBind_6lzg_COVID_hACE2_glycans_Zn_Lisinopril_NH_P1_BSite_100ns_All_Seeds.csv', 'Delta_MMGBSA_Lig_Specific_DBind_6lzg_COVID_hACE2_glycans_Zn_Lisinopril_P1_BSite_100ns_All_Seeds.csv', 'Delta_MMGBSA_Lig_Specific_DBind_6lzg_COVID_hACE2_glycans_Zn_Physcion_P1_BSite_100ns_All_Seeds.csv', 'Delta_MMGBSA_Lig_Specific_DBind_6lzg_COVID_hACE2_glycans_Zn_Physcion_P2_BSite_100ns_All_Seeds.csv', 
              'Delta_MMGBSA_Lig_Specific_DBind_6lzg_COVID_hACE2_glycans_Zn_Physcion_P3_BSite_100ns_All_Seeds.csv']
COM_txt = ['COM_Diquafosol_P1.txt', 'COM_Diquafosol_P3.txt', 'COM_Emodin_P1.txt', 'COM_Emodin_P2.txt', 'COM_Fosinopril_P1.txt', 'COM_Fosinoprilat_P1.txt', 'COM_Fosinoprilat_P2.txt', 'COM_Fosinoprilat_P3.txt', 'COM_Lisinopril_P1.txt', 'COM_Lisinopril_NH_P1.txt', 'COM_Physcion_P1.txt', 'COM_Physcion_P2.txt', 'COM_Physcion_P3.txt']

##################Uncomment below for new files. This allows the user to input the file names###################
# Number of files that contain MMGBSA and center of mass data, which will be used to calculate the percent dissociation

#MMGBSA text files
###MMGBSA_txt = []
###n = int(input("Enter the number of MMGBSA text files  : "))
###print("Now enter the names of", n, "MMGBSA text files.")
###print("Note that if the files are not in your current directory you will have to enter the pathways.")
# iterating till the range
###for i in range(0, n):
    ###file = str(input())
    # adding the file
    ###MMGBSA_txt.append(file)

#MMGBSA csv files
###MMGBSA_csv = []
###print("Now enter the names of", n, "MMGBSA csv files.") 
 
# iterating till the range
###for i in range(0, n):
    ###file = str(input())
    # adding the file
    ###MMGBSA_csv.append(file)
    
    
#Center of mass text files
###COM_txt = []
###print("Now enter the names of", n, "Center of Mass text files.")
 
# iterating till the range
###for i in range(0, n):
    ###file = str(input())
    # adding the file
    ###COM_txt.append(file)
    


###MAKING SURE IT READS EVERY LINE OF EVERY FILE
countL = []

#Looping the amount of poses/ligands for calculating the Percent Dissociation
for i in range(len(MMGBSA_csv)):
    
    #initialize counter for the percent dissociation. This is counting the number of frames.
    countPD = 0
    
    #this is a test: number of lines in a file ###comment out when done####
    countLines = 0
    
    #initialize a list that will keep track of the frames where the ligand is considered to no long be bound in the binding site
    Out_Frames = []
    AOF1 = []
    AOF2 = []
    
    #read the MMGBSA csv file from the list
    fileMMGBSA = MMGBSA_csv[i]
    
    #write to the MMGBSA txt file from the list 
    newMMGBSA = MMGBSA_txt[i]
    #print(fileMMGBSA)
    
    f1 = open(fileMMGBSA, 'r')
    fw = open(newMMGBSA, 'w')
    for line1 in f1:
        temp1 = line1.split()
        #if the line does not contain data or just contain a zero we need to skip all of these steps
        if len(temp1) > 1:
        
            #increase line counter ###COMMENT OUT WHEN TESTED###
            countLines = countLines +1
        
            #Writing a new MMGBSA file that only containes the ns time and the average binding estimation for future plots
            newLine = temp1[0] + '\t' + temp1[7] + '\n'
            fw.write(newLine)
        
        
            #When ever the MMGBSA indicates that the ligand is not binding increase frame counter (countPD) 
            #and take note of the frame in Out_Frames list.
            if float(temp1[7]) < 1.0 and float(temp1[7]) > -1.0:
                countPD = countPD+1
            
                #Column 1 is in ns, so convert ns to frame by multiplying by 10
                Out_Frames.append(float(temp1[0])*10)
                AOF1.append(float(temp1[0])*10)
    
    #Close the MMGBSA files       
    f1.close
    fw.close
    
    #read the COM file from the list
    fileCOM = COM_txt[i]
    f2 = open(fileCOM, 'r')
    for line2 in f2:
        temp2 = line2.split()
        
        #When the distance of ligand from the initial position is greater than or equal to 10 
        #and this frame was not accounted for in the previous MMGBSA section, increase frame counter
        #and take note of the frame in Out_Frames list.
        
        ###For a different length in Angstroms change the number in the first if statement
        if float(temp2[1]) >= 10.0:
            if float(temp2[0]) not in Out_Frames:
                countPD = countPD+1
                
                #Column 1 is already in frames
                Out_Frames.append(float(temp2[0]))
            AOF2.append(float(temp2[0]))
    
    #Close the COM file           
    f2.close
    
    #test appending countLines
    countL.append(countLines)
    if len(AOF1) == 0:
        AOF_MMGBSA.append([0])
    else:
        AOF_MMGBSA.append(AOF1)
    if len(AOF2) == 0:
        AOF_COM.append([0])
    else:
        AOF_COM.append(AOF2)
    
    #Calculating the final percent dissociations
    Final_PD.append(round((countPD/num_frames_MD_sims)*100,2)) 
    
    #Sorting the frames in ascending order and adding to a list
    Out_Frames.sort()
    All_Out_Frames.append(Out_Frames)
    
    
#Did it work?
#print(countL)
print("This is the percentage of dissociation")
print(Final_PD)
#print('\n')
#print("Frames where the ligand is not binding due to MMGBSA")
#print(AOF_MMGBSA)
#print("\n")
#print("Frames outside of the binding site, measuered by COM")
#print(AOF_COM)
#print(All_Out_Frames)
    
```
